```python
import pandas as pd
```

```python
dt = pd.read_csv('/Users/ayishahseimela/Downloads/car_prices.csv')
dt
```

```python

#remove null values 
dt = dt.dropna()
dt.isna().sum()
```

```python
#separate date and time 
#start by converting to datetime format
dt['saledate'] = pd.to_datetime(dt['saledate'], utc=True)

#create seperate date ans time column
dt['sale_date'] = dt['saledate'].dt.date
dt['sale_time'] = dt['saledate'].dt.time
dt = dt.drop(columns=['saledate'])
dt
```
