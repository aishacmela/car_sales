```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dt = pd.read_csv('/Users/ayishahseimela/Downloads/car_prices.csv')
dt
```

```python
dt['make'].isnull().mean() * 100
```

```python

dt["model"] = dt["model"].fillna(dt["model"].mode()[0])
print(dt["model"].isnull().sum())
```

```python

#remove null values 
dt = dt.dropna()
dt.isna().sum()
```

```python
#separate date and time 
#start by converting to datetime format
dt['saledate'] = pd.to_datetime(dt['saledate'], utc=True)

#create seperate date ans time column
dt['sale_date'] = dt['saledate'].dt.date
dt['sale_time'] = dt['saledate'].dt.time
dt = dt.drop(columns=['saledate'])
dt
```

```python
dt_sorted = dt.sort_values(
    by=['year', 'make'],
    ascending=[True, True]     
)
dt_sorted

year_count = dt['year'].value_counts().sort_index()
year_count.plot(
    kind= 'bar',
    figsize=(12,6),
    title='Number of Machines by year'
)
```

```python
#top 10 most common models 
top_models = dt['model'].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(top_models.index.astype(str),
        top_models.values)

plt.title('top 10 Most common models')
plt.xlabel('Model')
plt.ylabel('Frequency')

plt.tight_layout
plt.show()
```

```python
#number of unique models per year 

models_per_year = (
    dt.groupby('year')['model']
    .nunique()
)

plt.figure(figsize=(12,6))
plt.plot(models_per_year.index,
         models_per_year.values,
         marker='o')
plt.title('Unique model by year')
plt.xlabel('Year made')
plt.ylabel('Number of unique models')
plt.grid(True)
plt.show()
```
